```python
from datasets.clean_noisy_dataset import SIDDDataset
from torchvision import transforms

ds = SIDDDataset("./SIDD_Small_sRGB", transforms.Compose([
#     transforms.RandomRotation((-180, 180)),
    transforms.RandomCrop(1024),
    transforms.RandomHorizontalFlip(0.5),
    transforms.RandomVerticalFlip(0.5),
    transforms.ToTensor()
]))
```

```python
first = ds[0]
clean = first['clean']
noisy = first['noisy']
clean.shape, noisy.shape
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.imshow(noisy.permute(1,2,0))
ax2.imshow(clean.permute(1,2,0))
```
